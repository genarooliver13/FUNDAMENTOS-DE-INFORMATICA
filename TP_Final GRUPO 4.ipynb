```python
#1
import pandas as pd
import math
import datetime


# ============= FUNCIONES =====================

def dimensionar_zapata(carga_columna, punto_muestreo, profundidad_deseada, forma, relacion_lados, archivo_csv):
    """
    Dimensiona una zapata seg√∫n la carga y tipo de suelo.
    """
    df = pd.read_csv(archivo_csv)
    df_p = df[df['punto_muestreo'] == punto_muestreo]
    if df_p.empty:
        print(f"‚ö†Ô∏è Punto {punto_muestreo} no encontrado en el CSV.")
        return None

    # Tomar la fila m√°s cercana a la profundidad deseada
    fila = df_p.iloc[(df_p['profundidad_m'] - profundidad_deseada).abs().argsort().iloc[0]]
    resistencia_suelo = fila['resistencia_kpa']  # kPa = kN/m¬≤
    tipo_suelo = fila['tipo_suelo']

    # C√°lculo del √°rea necesaria
    area = carga_columna / resistencia_suelo  # m¬≤

    if forma == 'cuadrada':
        lado = math.sqrt(area)
        dimensiones = f"Lado = {lado:.2f} m"
    elif forma == 'rectangular':
        ancho = math.sqrt(area / relacion_lados)
        largo = ancho * relacion_lados
        dimensiones = f"Largo = {largo:.2f} m, Ancho = {ancho:.2f} m"
    else:
        return "‚ö†Ô∏è Forma inv√°lida"

    print(f"\nüß± Resultados del dimensionamiento de zapata ({punto_muestreo}):")
    print(f"Tipo de suelo: {tipo_suelo}")
    print(f"Resistencia: {resistencia_suelo} kPa")
    print(f"√Årea requerida: {area:.2f} m¬≤")
    print(dimensiones)

    return {
        "tipo_suelo": tipo_suelo,
        "resistencia_suelo": resistencia_suelo,
        "area": area,
        "dimensiones": dimensiones
    }


def profundidad_minima_para_resistencia(punto_muestreo, resistencia_objetivo, archivo_csv):
    """
    Busca la profundidad m√≠nima que alcanza la resistencia deseada.
    """
    df = pd.read_csv(archivo_csv)
    df_p = df[df['punto_muestreo'] == punto_muestreo]

    if df_p.empty:
        print("‚ö†Ô∏è Punto no encontrado en el archivo.")
        return None

    df_filtrado = df_p[df_p['resistencia_kpa'] >= resistencia_objetivo]
    if df_filtrado.empty:
        print(f"‚ö†Ô∏è En {punto_muestreo} no se alcanza {resistencia_objetivo} kPa.")
        return None

    profundidad_min = df_filtrado['profundidad_m'].min()
    return profundidad_min


def humedad_promedio_por_punto(archivo_csv):
    """
    Calcula la humedad promedio de cada punto de muestreo.
    """
    df = pd.read_csv(archivo_csv)
    return df.groupby('punto_muestreo')['humedad_porcentaje'].mean().to_dict()


def generar_informe(archivo_csv, punto, resistencia_deseada, carga_columna, profundidad_deseada, forma, relacion_lados):
    """
    Genera informe de resultados y lo guarda como .txt
    """
    profundidad_min = profundidad_minima_para_resistencia(punto, resistencia_deseada, archivo_csv)
    humedad_prom = humedad_promedio_por_punto(archivo_csv)
    zapata = dimensionar_zapata(carga_columna, punto, profundidad_deseada, forma, relacion_lados, archivo_csv)

    fecha = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    nombre_informe = f"informe_{punto}.txt"

    with open(nombre_informe, "w") as f:
        f.write("INFORME DE RESULTADOS - ENSAYOS DE SUELO\n")
        f.write(f"Fecha de generaci√≥n: {fecha}\n\n")
        f.write(f"Punto de muestreo: {punto}\n")
        f.write(f"Resistencia deseada: {resistencia_deseada} kPa\n")
        f.write(f"Profundidad m√≠nima necesaria: {profundidad_min} m\n\n")
        f.write(f"Humedades promedio:\n")
        for p, h in humedad_prom.items():
            f.write(f" - {p}: {h:.2f}%\n")
        f.write("\n--- Dimensionamiento de zapata ---\n")
        if zapata:
            f.write(f"Tipo de suelo: {zapata['tipo_suelo']}\n")
            f.write(f"Resistencia suelo: {zapata['resistencia_suelo']} kPa\n")
            f.write(f"√Årea requerida: {zapata['area']:.2f} m¬≤\n")
            f.write(f"Dimensiones: {zapata['dimensiones']}\n")

    print(f"\nüìÑ Informe generado: {nombre_informe}")
    return nombre_informe


# ============= PROGRAMA INTERACTIVO =====================

print("üîπ Bienvenido al sistema de gesti√≥n de proyectos - Nivel 2")
punto = input("Ingrese el punto de muestreo (ej: P1 a P7): ").strip().upper()
resistencia_deseada = float(input("Ingrese la resistencia deseada (kPa): "))
carga_columna = float(input("Ingrese la carga de la columna (kN): "))
profundidad_deseada = float(input("Ingrese la profundidad deseada (m): "))
forma = input("Ingrese la forma de la zapata ('cuadrada' o 'rectangular'): ").strip().lower()
relacion_lados = float(input("Ingrese la relaci√≥n largo/ancho (1.0 si es cuadrada): "))

# Ejecutar todo y generar informe
archivo_csv = "ensayos_suelo.csv"
informe = generar_informe(
    archivo_csv=archivo_csv,
    punto=punto,
    resistencia_deseada=resistencia_deseada,
    carga_columna=carga_columna,
    profundidad_deseada=profundidad_deseada,
    forma=forma,
    relacion_lados=relacion_lados
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# 1) CARGA DEL CSV ORIGINAL
# ===============================
df = pd.read_csv(
    '/content/Copia de prestamos-de-la-biblioteca-municipal-2022-2024.csv',
    sep=';',
    engine='python',
    encoding='latin1',
    on_bad_lines='skip'
)

# ===============================
# 2) LIMPIEZA B√ÅSICA
# ===============================
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.columns = df.columns.str.encode('latin1', errors='ignore').str.decode('utf-8', errors='ignore')

print("Columnas detectadas:", df.columns.tolist())

# ===============================
# 3) DETECCI√ìN AUTOM√ÅTICA DE COLUMNAS
# ===============================

# Fecha
fecha_col = next(col for col in df.columns if "fecha" in col.lower())
# T√≠tulo
titulo_col = next(col for col in df.columns if "tul" in col.lower())
# Autor
autor_col = next(col for col in df.columns if "autor" in col.lower())

print("\nColumna Fecha   ‚Üí", fecha_col)
print("Columna T√≠tulo  ‚Üí", titulo_col)
print("Columna Autor   ‚Üí", autor_col)

df.rename(columns={
    fecha_col: "Fecha Operaci√≥n",
    titulo_col: "T√≠tulo",
    autor_col: "Autor Principal"
}, inplace=True)

# ===============================
# 4) PROCESAR FECHAS
# ===============================
df["Fecha Operaci√≥n"] = pd.to_datetime(df["Fecha Operaci√≥n"], errors="coerce")
df = df.dropna(subset=["Fecha Operaci√≥n"])
df["A√±o"] = df["Fecha Operaci√≥n"].dt.year

# ===============================
# 5) RANKING DE LIBROS POR A√ëO
# ===============================
print("\n==============================")
print("  RANKING DE LIBROS POR A√ëO")
print("==============================\n")

for a√±o in sorted(df["A√±o"].unique()):
    print(f"A√±o {a√±o}:")
    print(df[df["A√±o"] == a√±o]["T√≠tulo"].value_counts().head(10))
    print()

# ===============================
# 6) LIBROS DISTINTOS
# ===============================
print("==============================")
print("  CANTIDAD DE LIBROS DISTINTOS")
print("==============================")
print("Libros distintos:", df["T√≠tulo"].nunique(), "\n")

# ===============================
# 7) AUTOR CON M√ÅS LIBROS
# ===============================
print("==============================")
print("  AUTOR M√ÅS PRESTADO")
print("==============================")
print(df["Autor Principal"].value_counts().head(10), "\n")

# ===============================
# 8) GR√ÅFICOS (MOSTRAR + GUARDAR)
# ===============================

# -- Distribuci√≥n por edad --
plt.figure(figsize=(8,5))
df["Edad"].dropna().astype(int).hist(bins=20)
plt.title("Distribuci√≥n por Edad")
plt.xlabel("Edad")
plt.ylabel("Cantidad")
plt.show()
plt.savefig("/content/grafico_edad.png")
plt.close()

# -- Pr√©stamos por d√≠a --
plt.figure(figsize=(8,5))
df["Fecha Operaci√≥n"].dt.day_name().value_counts().plot(kind="bar")
plt.title("Pr√©stamos por D√≠a de la Semana")
plt.xlabel("D√≠a")
plt.ylabel("Cantidad")
plt.show()
plt.savefig("/content/grafico_dias.png")
plt.close()

# -- Pr√©stamos por mes --
plt.figure(figsize=(8,5))
df["Fecha Operaci√≥n"].dt.month.value_counts().sort_index().plot(kind="bar")
plt.title("Pr√©stamos por Mes")
plt.xlabel("Mes")
plt.ylabel("Cantidad")
plt.show()
plt.savefig("/content/grafico_meses.png")
plt.close()

# -- Pr√©stamos por a√±o --
plt.figure(figsize=(8,5))
df["A√±o"].value_counts().sort_index().plot(kind="bar")
plt.title("Pr√©stamos por A√±o")
plt.xlabel("A√±o")
plt.ylabel("Cantidad")
plt.show()
plt.savefig("/content/grafico_a√±os.png")
plt.close()

print("‚úì Listo: an√°lisis terminado y gr√°ficos generados.")
```
